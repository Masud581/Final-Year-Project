```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset from Google Colab (assuming you have uploaded the file using the Colab interface)
from google.colab import files
uploaded = files.upload()

# Read the Excel file (change the filename to the actual uploaded file name)
data = pd.read_excel(list(uploaded.keys())[0])

# Set up the plotting theme
sns.set(style="whitegrid")

# Check the first few rows of the dataset to ensure correct loading
print(data.head())

# List of numeric columns to visualize
numeric_columns = ['Age', 'PlayTimeHours', 'SessionsPerWeek', 'AvgSessionDurationMinutes',
                   'PlayerLevel', 'AchievementsUnlocked']

# 1. Plot histograms for all numeric columns
for col in numeric_columns:
    if col in data.columns:
        plt.figure(figsize=(10, 6))
        sns.histplot(data[col], kde=True, bins=30, color='blue')
        plt.title(f'Distribution of {col}', fontsize=15)
        plt.xlabel(col, fontsize=12)
        plt.ylabel('Frequency', fontsize=12)
        plt.show()

# List of categorical columns to visualize
categorical_columns = ['Gender', 'Location', 'GameGenre', 'GameDifficulty', 'EngagementLevel']

# 2. Histograms for Age vs categorical variables (grouped)
for cat_col in categorical_columns:
    if 'Age' in data.columns and cat_col in data.columns:
        plt.figure(figsize=(10, 6))
        sns.histplot(data=data, x='Age', hue=cat_col, multiple="stack", bins=30)
        plt.title(f'Age Distribution Grouped by {cat_col}', fontsize=15)
        plt.xlabel('Age', fontsize=12)
        plt.ylabel('Count', fontsize=12)
        plt.show()

# 3. Histograms for PlayerLevel vs categorical variables (grouped)
for cat_col in categorical_columns:
    if 'PlayerLevel' in data.columns and cat_col in data.columns:
        plt.figure(figsize=(10, 6))
        sns.histplot(data=data, x='PlayerLevel', hue=cat_col, multiple="stack", bins=30)
        plt.title(f'Player Level Distribution Grouped by {cat_col}', fontsize=15)
        plt.xlabel('Player Level', fontsize=12)
        plt.ylabel('Count', fontsize=12)
        plt.show()

# 4. Histograms for PlayTimeHours vs categorical variables (grouped)
for cat_col in categorical_columns:
    if 'PlayTimeHours' in data.columns and cat_col in data.columns:
        plt.figure(figsize=(10, 6))
        sns.histplot(data=data, x='PlayTimeHours', hue=cat_col, multiple="stack", bins=30)
        plt.title(f'Play Time Hours Distribution Grouped by {cat_col}', fontsize=15)
        plt.xlabel('Play Time Hours', fontsize=12)
        plt.ylabel('Count', fontsize=12)
        plt.show()

# 5. Countplot for categorical vs categorical variables
for cat_col1 in categorical_columns:
    for cat_col2 in categorical_columns:
        if cat_col1 != cat_col2 and cat_col1 in data.columns and cat_col2 in data.columns:
            plt.figure(figsize=(10, 6))
            sns.countplot(x=cat_col1, hue=cat_col2, data=data)
            plt.title(f'{cat_col1} vs {cat_col2}', fontsize=15)
            plt.xlabel(cat_col1, fontsize=12)
            plt.ylabel('Count', fontsize=12)
            plt.show()

# 6. Age vs Achievements Unlocked
if 'Age' in data.columns and 'AchievementsUnlocked' in data.columns:
    plt.figure(figsize=(10, 6))
    sns.histplot(data=data, x='Age', y='AchievementsUnlocked', bins=30, pthresh=.1, cmap="mako")
    plt.title('Age vs Achievements Unlocked', fontsize=15)
    plt.xlabel('Age', fontsize=12)
    plt.ylabel('Achievements Unlocked', fontsize=12)
    plt.show()

# 7. Age vs PlayerLevel
if 'Age' in data.columns and 'PlayerLevel' in data.columns:
    plt.figure(figsize=(10, 6))
    sns.histplot(data=data, x='Age', y='PlayerLevel', bins=30, pthresh=.1, cmap="coolwarm")
    plt.title('Age vs Player Level', fontsize=15)
    plt.xlabel('Age', fontsize=12)
    plt.ylabel('Player Level', fontsize=12)
    plt.show()

# 8. Location vs In-Game Purchases (Countplot)
if 'Location' in data.columns and 'InGamePurchases' in data.columns:
    plt.figure(figsize=(10, 6))
    sns.countplot(x='Location', hue='InGamePurchases', data=data)
    plt.title('Location vs In-Game Purchases', fontsize=15)
    plt.xlabel('Location', fontsize=12)
    plt.ylabel('Count', fontsize=12)
    plt.show()

# 9. Gender vs Game Difficulty (Countplot)
if 'Gender' in data.columns and 'GameDifficulty' in data.columns:
    plt.figure(figsize=(10, 6))
    sns.countplot(x='Gender', hue='GameDifficulty', data=data)
    plt.title('Gender vs Game Difficulty', fontsize=15)
    plt.xlabel('Gender', fontsize=12)
    plt.ylabel('Count', fontsize=12)
    plt.show()

# 10. Age vs GameGenre (Histogram)
if 'Age' in data.columns and 'GameGenre' in data.columns:
    plt.figure(figsize=(10, 6))
    sns.histplot(data=data, x='Age', hue='GameGenre', multiple="stack", bins=30)
    plt.title('Age Distribution by Game Genre', fontsize=15)
    plt.xlabel('Age', fontsize=12)
    plt.ylabel('Count', fontsize=12)
    plt.show()
```
